```python
# importing libraries

import numpy as np
import pandas as pd
from simple_ml.simple_nn import MLPClassifier # our custom package for MLPClassifier
from sklearn.datasets import make_classification, load_digits
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# EXAMPLE 1

# testing with binary classification dataset

X, y = make_classification(n_samples=2000, random_state=47)
print(f'X shape: {X.shape}')
print(f'y shape: {y.shape}')
print(f'X sample: {X[:5]}')
print(f'y sample: {y[:5]}')
```

```python
# splitting input features and target labels into train and test data

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=47)
print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'y_test shape: {y_test.shape}')
```

```python
# creating a classifier and fitting with training data

clf1 = MLPClassifier(epochs=300)
clf1.fit(X_train, y_train)
```

```python
# predicting on sample test data

print(f'Actual: {y_test[:10]}')
print(f'Predicted: {clf1.predict(X_test[:10, :])}')
```

```python
# overall accuracy on whole test data

print(f'Test accuracy: {clf1.score(X_test, y_test)}')
```

```python
# plotting the loss curve while training

clf1.plot_loss_curve()
```

```python
# EXAMPLE 2

# testing with multiclass classification dataset (MNIST) 

mnist = load_digits()
X, y = mnist.data, mnist.target
print(f'X shape: {X.shape}')
print(f'y shape: {y.shape}')
```

```python
# sample feature data

pd.DataFrame(mnist.data).head()
```

```python
# sample label data

pd.DataFrame(mnist.target).head()
```

```python
# plotting sample images from MNIST dataset

fig, axes = plt.subplots(2, 10, figsize=(16, 6))
for i in range(20):
    axes[i//10, i %10].imshow(mnist.images[i], cmap='gray');
    axes[i//10, i %10].axis('off')
    axes[i//10, i %10].set_title(f"target: {mnist.target[i]}")
plt.tight_layout()
```

```python
# splitting input features and target labels into train and test data

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=47)
print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'y_test shape: {y_test.shape}')
```

```python
# creating a classifier and fitting with training data

clf2 = MLPClassifier(hidden_layer_sizes=(100, 100, 100))
clf2.fit(X_train, y_train)
```

```python
# predicting on sample test data

print(f'Actual: {y_test[:10]}')
predicted = clf2.predict(X_test[:10, :])
print(f'Predicted: {predicted}')
fig, axes = plt.subplots(1, 10, figsize=(16, 6))
for i in range(10):
    axes[i].imshow(X_test[i, :].reshape(8,8), cmap='gray');
    axes[i].axis('off')
    axes[i].set_title(f"target: {y_test[i]}\npredicted: {predicted[i]}")
plt.tight_layout()
```

```python
# overall accuracy on whole test data

print(f'Test accuracy: {clf2.score(X_test, y_test)}')
```

```python
# plotting the loss curve while training

clf2.plot_loss_curve()
```
